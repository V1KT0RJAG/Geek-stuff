##### Aufgabe B1: DEMO Series-Objekt

Series ist ein eindimensionales listenartiges Datenobjekt. Die folgenden Zellen demonstrieren einige grundlegende Eigenschaften der Klasse. 

https://pandas.pydata.org/docs/reference/series.html

````python
# Nicht löschen - Teil der Aufgabe
import pandas as pd

data1 = [10,20,30,40,50]
s1 = pd.Series(data=data1, index=['a', 'b', 'c', 'd', 'e'], name='mydata1')
# Die Elemente im Index werden auch als Labels bezeichnet.

data2 = {'a':1, 'b':2, 'c':3, 'd':5, 'f':6}
s2 = pd.Series(data2, name='mydata2')
````

````python
s2
````

Beispiele für Attribute des Series-Objektes

````python
# Nicht löschen - Teil der Aufgabe
print(s1.name)		# Name
print(s1.index)		# Index
print(s1.dtypes)	# Datentyp
print(s1.size) 		# Anzahl der Element
print(s1.shape)		# Länge als Tuple
````

Beispiele für Methoden des Series-Objektes

````python
# Nicht löschen - Teil der Aufgabe
s1.mean() # Berechnung des Mittelwertes
s1
````

````python
# Nicht löschen - Teil der Aufgabe
#isin checks if the values are found in the frame
s1.isin([15,10,13,40])
````

````python
# Nicht löschen - Teil der Aufgabe
# Index des größten Elements
s1.idxmax()
````

````python
# Nicht löschen - Teil der Aufgabe
s1.memory_usage() # Ausgabe des Speicherbedarfs
````

````python
# Nicht löschen - Teil der Aufgabe
s1.astype(float) # Ändern des Datentypes
````

Methoden geben typischerweise ein neues Objekt zurück und haben nicht ohne weiteres eine Wirkung auf das Objekt selbst!

````python
# Nicht löschen - Teil der Aufgabe
s1.rename('data')
````

````python
# Nicht löschen - Teil der Aufgabe
# rename hat keine Auswirkung auf s1
s1
````

Zugriff auf Elemente oder Teilmengen der Daten (Indexing)

-	Label indexing,

### D) Pandas: Gruppieren und Aggregieren (DEMO)

Sie haben bereits den mittleren Preise für ein einzelnes Modell oder mehrere Modelle berechnet.

Die Methode `groupby` vereinfacht solche Problemstellungen. Mit ihr können Sie Daten nach den Werten einer Spalte gruppieren, um z.B. Berechnungen für jede dieser Gruppen durchzuführen. Schauen wir uns jedoch vorerst an, was `groupby` bewirkt.

#### Aufgaben D1: DEMO GROUPBY-Objekte - Was bewirkt `groupby`?

````python
# Teil der Aufgabe nicht löschen
# Dieses Aufruf gibt ein GroupBy-Objekt zurück
type(df.groupby(['gear']))
````

Analysieren und verstehen Sie den folgenden Code!

````python
# Teil der Aufgabe nicht löschen
# Über das Groupby-Objekt (pandas.core.groupby.generic.DataFrameGroupBy) kann man iterieren.
# Es enthält den jeweiligen Wert der gruppierten Spalte und einen Dataframe
grouby_obj = df[['make','model','gear','price']].groupby('gear')
for i,n in enumerate(grouby_obj):
    print('Gruppe:',i)
    print(' Type:',type(n))
    print(' 1.Element des Tuples:\n',type(n[0]))#,n[0])
    print(' 2.Element des Tuples:\n',type(n[1]))#,n[1])
    print('     Spalten:',n[1].columns)#,n[1])
    x = n[1]
x
````

Ein GroupBy-Objekt stellt im Wesentlichen die selben Methoden zur Verfügung wie ein DataFrame. Daher kann man mit ihm wie mit einem DataFrame arbeiten.

````python
# Teil der Aufgabe nicht löschen
# Berechnung der Mittelwerte für die Gruppen in gear
df.groupby('gear').mean(numeric_only=True)
````

````python
# Teil der Aufgabe nicht löschen
# Berechnung des Mittelwertes für den Preis
df.groupby('gear')['price'].mean(numeric_only=True)
````

Gruppieren nach mehreren Spalten und Multi-Index

````python
# Teil der Aufgabe nicht löschen
# Gruppierung nach mehreren Kriterien/Spalten
# Es wird ein Multi-Index erzeugt
df.groupby(['fuel','gear'])['price'].mean(numeric_only=True)#.rename('mean_price').unstack()
````

````python
# Teil der Aufgabe nicht löschen
# Auflösen des Multi-Index
# Die Spalte mit dem mittleren Preis heißt jetzt Preis!
df.groupby(['fuel','gear'])['price'].mean(numeric_only=True).reset_index()#.rename(columns={'price':'mean_price'})
````

##### Aufgabe D2: Aggreationen mit `agg`

Mehrere Aggregationen für die gleichen Spalten mit `agg`oder `aggregate`

````python
# Teil der Aufgabe nicht löschen
df.agg('mean', numeric_only=True) # entspricht df.mean(numeric_only=True)
````

````python
# Teil der Aufgabe nicht löschen
# Spaltennamen des Resultats richten sich nach den verwendeten Funktionen
df.groupby(['fuel','gear'])['price'].agg(['mean', 'min', 'max','count'])#.reset_index()
````

Verschiedene Aggregationen für verschiedene Spalten

````python
# Teil der Aufgabe nicht löschen
df.groupby(['fuel','gear']).agg(
    {
        'hp':'nunique',
        'price':['mean', 'min', 'max']    
    }
)
````

Verschiedene Aggregationen für verschieden Spalten und Umbenennen der Spalten

````python
# Teil der Aufgabe nicht löschen
df.groupby(['fuel','gear']).agg(
    count=('price','count'),
    mean_price=('price','mean'),
    hp_uniques=('hp','nunique')
)#.unstack()
````

Eigene Funktionen für die Aggregation definieren

````python
# Teil der Aufgabe nicht löschen
# Argument ist Series der Gruppe
# Rückgabe ist int -> erscheint in Aggregation
def f(s:pd.Series)->bool:
    return (s=='Ford').sum()

df.groupby(['fuel','gear']).agg(
    count=('price','count'),    # Built-in Aggregation count
    count_ford=('make',f),      # eigene Definition mittels Funtion    
    mean_price=('price','mean'),# Built-in Aggregation mean
    hp_uniques=('hp','nunique') # Built-in Aggregation nunique
)
````
